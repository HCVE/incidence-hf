```python
from hcve_lib.visualisation import h2


def show_hyperparameter_analysis(run_id: str):
    with open(get_hyperparameters_dependency_file(run_id), 'rb') as f:
        analysis_data = pickle.load(f)

    run_info = get_run_info(run_id)

    h2(f'{run_info["run_name"]} ({run_info["experiment_name"]})')

    for split_name, split_data in analysis_data.items():
        b(split_name)

        scatter_data = {
            'inner': split_data['c_indexes_inner'],
            'outer': split_data['c_indexes_outer'],
            'hyperparameters':  split_data['hyperparameters'],
        }

        fig = px.scatter(
            scatter_data,
            x='inner',
            y='outer',
            hover_data={'inner': ':.3f', 'outer': ':.3f', 'hyperparameters': True},
            width=400,
            height=400,
            color_discrete_sequence=px.colors.qualitative.Pastel,
            template='simple_white',
        )
        fig.update_traces(marker=dict(size=3))
        fig.update_layout(
            xaxis_title="Inner c-index",
            yaxis_title="Outer c-index",
            legend = {
                'title': {'text': None}
            },
        )
        fig.update_xaxes(
            range = [0.5, 1]
          )

        fig.update_yaxes(
            scaleanchor = "x",
            scaleratio = 1,
            range = [0.5, 1]
         )

        fig.show()
        display_html('<br>')
```
